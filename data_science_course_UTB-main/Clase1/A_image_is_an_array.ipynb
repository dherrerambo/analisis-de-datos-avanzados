```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
%matplotlib inline
```

```python
img = cv.imread('universe-eyes-nebula-helix-nebula-wallpaper.jpg')
```

```python
print("The shape of image is =",np.shape(img))
```

```python
plt.figure(figsize=(8,9))
plt.imshow(img)
plt.show()
```

```python
from skimage import io, color
```

```python
img2=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(8,9))
plt.imshow(img2)
plt.show()
```

```python
img3=cv.cvtColor(img2,cv.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(8,9))
plt.imshow(img3)
plt.show()
```

```python
r,g,b=cv.split(img2)
```

```python
plt.figure(figsize=(28,19))
plt.subplot(131)
im=plt.imshow(b,cmap="Blues")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.subplot(132)
im=plt.imshow(r,cmap="Reds")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.subplot(133)
im=plt.imshow(g,cmap="Greens")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(b.ravel(),bins=256,range=(0,256),color='b',label='blue')
plt.hist(g.ravel(),bins=256,range=(0,256),color='g',label='green')
plt.hist(r.ravel(),bins=256,range=(0,256),color='r',label='red')
plt.legend()
plt.show()
```

```python
img = cv.imread('Screenshot.png')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim(0,256)
plt.show()
```
